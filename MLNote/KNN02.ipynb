# 카테고리 변수 처리

```python
col_list=[]
for i in data.columns:
  if data[i].dtype == 'O':
    col_list.append(i)
```

```python
col_list
```

```python
for i in col_list:
  print(i, data[i].nunique())
```

```python
col_list = col_list[1:]
```

```python
col_list
```

```python
data = pd.get_dummies(data, columns=col_list, drop_first=True)
```

```python
data
```

# 결측치 (Missing Value) 처리

```python
data.isna().sum()
```

```python
data['TotalCharges'] = data['TotalCharges'].fillna(data['TotalCharges'].median())
```

```python
data.isna().sum()
```

```python
data
```

# Scaling

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler, RobustScaler
```

### Min-Max Scaler

```python
minmax = MinMaxScaler()
```

```python
data
```

```python
data.drop('customerID', axis=1, inplace=True)
```

```python
minmax.fit(data)
```

```python
scaled_data = minmax.transform(data)
```

```python
scaled_data
```

```python
pd.DataFrame(scaled_data, columns=data.columns)
```

### 최종적으로 Min-Max 스케일러로 데이터 저장

```python
scaled_data = pd.DataFrame(scaled_data, columns=data.columns)
```

# Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X = scaled_data.drop('Churn_Yes', axis=1)
y = scaled_data['Churn_Yes']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=100)
```

# Modeling

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=10)
```

```python
knn.fit(X_train, y_train)
```

```python
pred = knn.predict(X_test)
```

```python
pd.DataFrame({'acutal_value' : y_test, 'pred_value' : pred}).head(30)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, pred)
```

#  최적의 K값 구하기 (튜닝)

```python
accuracy_list = []
```

```python
for i in range(1, 101):
  knn = KNeighborsClassifier(n_neighbors= i)
  knn.fit(X_train, y_train)
  pred = knn.predict(X_test)
  accuracy_list.append(accuracy_score(y_test, pred))
```

```python
accuracy_list
```

```python
plt.figure(figsize=(20, 10))
sns.lineplot(x=range(1, 101), y = accuracy_list, marker='o', markersize=10, markerfacecolor='red')
```
