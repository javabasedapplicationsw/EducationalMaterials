### groupby()
  - 괄호 안에 변수 이름 넣음
  - 추가 연산을 붙여야 함
    - sum(), mean(), std() 함수
    

```python
data.groupby('native-country').mean().sort_values('class')
# 그룹별 평균 계산 후 class 기준으로 오름차순 정렬
```

```python
country_group = data.groupby('native-country').mean()['class']   # native-country별로 class의 평균값을 계산하여 저장
```

```python
country_group
```

```python
# reset_index() : 인덱스를 변수로 불러냄
country_group = country_group.reset_index()
```

```python
country_group
```

```python
# data와 country_group를 native-country 기준으로 결합 (left 조인)
data = data.merge(country_group, on= 'native-country', how= 'left')
```

```python
# rename() : 데이터프레임의 변수나 인덱스 이름을 변경할수 있음.  변경하려는 이름과 변경 후의 이름을 딕셔너리 형태로 넣어주면 됨
data.drop('native-country', axis=1, inplace=True)   # 변수 제거
data = data.rename(columns= {'class_x': 'class', 'class_y': 'native-country'})    # 변수 이름 변환
```

```python
data
```

## 전처리 : 결측치 처리 및 더미 변수 변환

```python
data.isna().mean()      # 결측치 비율 확인
```

```python
# native-country
# 보통 결측치를 숫자로 채워줄 때 -9나 -99와 임의의 숫자를 사용함
# - 트리 기반 모델에서는 큰 문제가 없음, 선형 모델에서는 데이터의 왜곡 발생 할수 있으니 주의해야 함

data['native-country'] = data['native-country'].fillna(-99)       # 결측치를 -99로 대체
```

```python
data.isna().mean()      # 결측치 비율 확인
```

```python
# workclass은 범주형 변수이므로 평균치로 해결할수 없음
# - 이 경우는 특정 텍스트로 채워주거나, 제거해야 함
data['workclass'].value_counts()       # 고유값별 출현 빈도 확인
```

```python
data['workclass'] = data['workclass'].fillna('Private')    # 결측치를 Private으로 대체
```

```python
data.isna().mean()      # 결측치 비율 확인
```

```python
# occupation 컬럼의 고윳값별 출현 빈도 확인
data['occupation'].value_counts()
```

```python
# 결측치를 Unknown으로 대처
data['occupation'] = data['occupation'].fillna('Unknown')
```

```python
data.isna().mean()      # 결측치 비율 확인
```

```python
# 더미 변수로 변환
data = pd.get_dummies(data, drop_first=True)
```

```python
data
```

### 모델링 및 평가하기

```python
# 데이터를 훈련셋과 시험셋으로 나누기
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data.drop('class', axis=1), data['class'], test_size= 0.4, random_state = 100)   # 훈련셋/시험셋 분리
```

```python
from sklearn.tree import DecisionTreeClassifier    # 0과 1을 예측해야 함 (분류모델은 범주형 변수용)
```

```python
# 모델 객체 생성
model = DecisionTreeClassifier()
```

```python
# 학습
model.fit(X_train, y_train)
```

```python
# 예측
pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, pred)   # 정확도 계산 (81%)
```
