```python
# Name, Ticket 두 변수는 더미 변수로 변환시키면 그 수만큼 컬럼이 생김
# Name, Ticket 두 변수를 데이터에서 제외하고, 남은 Sex, Embarked은 원-핫 인코딩 적용
data = data.drop(['Name', 'Ticket'], axis=1)
```

```python
data
```

```python
# get_dummies() 함수 사용
# - 문자 형태의 변수들을 원-핫-인코딩함
# ==> 총 5개의 컬럼이 오른쪽에 생성됨, 기존에 있던 컬럼2개는 사라짐
pd.get_dummies(data, columns=['Sex', 'Embarked'])
```

```python
pd.get_dummies(data, columns= ['Sex', 'Embarked'], drop_first=True)
# 첫번째 항목 제거됨, 총 3개의 더미 변수가 새로 생김 ==> 계산량이 그만큼 줄어듦
# 10개 ==> 8개 컬럼이 줄어듦
```

```python
data = pd.get_dummies(data, columns= ['Sex', 'Embarked'], drop_first=True)
```

```python
data
```

```python
# 독립변수, 종속변수, 훈련셋, 시험셋으로 나누기
from sklearn.model_selection import train_test_split

X = data.drop('Survived', axis=1)   # 데이터셋에서 종속변수 제거 후 저장
y = data['Survived']     # 데이터셋에서 종속변수만 저장
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.2, random_state=100)  # 학습셋, 실험셋 분리
```

```python
X_train
```

```python
X_test
```

```python
# 로지스틱 회귀
from sklearn.linear_model import LogisticRegression
```

```python
# 로지스틱 회귀 모델 생성
model = LogisticRegression()

# 모델 학습
model.fit(X_train, y_train)
```

```python
# 예측하기
pred = model.predict(X_test)
```

```python
y_test
```

```python
# 정확도 라이브러리 임포트
from sklearn.metrics import accuracy_score
# 실제값과 예측값으로 정확도 계산
accuracy_score(y_test, pred)
```

```python
# 위는 약 78% 정도의 정확도를 보임
```

```python
# 피처 엔지니어링 (Feature Engineering)
# SibSp와 Parch 변수 합치기
data['family'] = data['SibSp'] + data['Parch']
data.drop(['SibSp', 'Parch'], axis=1, inplace=True)       # SibSp와 Parch 변수 삭제
data.head()         # 5행 출력
```

```python
X = data.drop('Survived', axis=1)
y = data['Survived']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=100) # 학습셋, 시험셋 준비
model = LogisticRegression()      # 모델 생성
model.fit(X_train, y_train)       # 학습
pred = model.predict(X_test)      # 예측
accuracy_score(y_test, pred)      # 정확도 계산
```
