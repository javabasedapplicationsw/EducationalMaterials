```python
import pandas as pd
```

### 데이터프레임 합치기
  - 내부 조인 (INNER JOIN)
    - 양쪽 테이블에 공통된 키값에 대해서만 데이터 결합
  - 왼쪽 조인 (LEFT JOIN)
    - 왼쪽 테이블에 있는 키값을 기준으로 오른쪽 테이블을 가져다 붙이는 형태
  - 오른쪽 조인 (RIGHT JOIN)
    - 오른쪽 테이블에 있는 키값을 기준으로 왼쪽 테이블을 가져다 붙이는 형태
  - 전체 조인 (FULL JOIN)
    - 공통되지 않은 키값까지 모두 포함해 데이터 결합

```python
left_url = 'left.csv'
right_url = 'right.csv'
left = pd.read_csv(left_url)
right = pd.read_csv(right_url)
```

### key에 주목하기
  - key값을 기준으로 데이터를 합쳐줄 것임
  - 겹치는 key는 b,c,e 임

```python
# merge()함수로 두 테이블을 겹합하기
# 기본적으로 내부 조인 수행함 => 두 테이블에 공통된 key인 b,c,e만 겹합되고, 나머지는 버려짐
# merge()함수는 자동적으로 두 테이블에 공통으로 존재하는 컬럼명을 찾아 키값으로 활용함
left.merge(right)
```
