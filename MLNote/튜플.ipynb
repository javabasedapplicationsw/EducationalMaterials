# 자료구조 - 시퀀스 (sequence)


### 요소 (element)로 구성, 요소 간에는 순서가 있음 ,
### 내장 시퀀스
  - str, list, tuple, range

### 동일한 연산을 지원
  - 인덱싱(indexing), 슬라이싱(slicing), 덧셈 연산(adding), 곱셈 연산(multiplying)
  - 내장 함수 적용가능   

## 튜플
- 리스트와 아주 유사함
- 리스트와 다르게 튜플은 변경이 불가능 함
- 형식
  - 튜플 이름 = ( 항목1, 항목2, ...)

```python
fruits = ("apple", "banana", "grace")
```

```python
fruits
```

```python
single_tuple = ("apple",)      # 쉼표가 끝에 있어야 함
```

```python
single_tuple
```

```python
type(single_tuple)
```

```python
single_tuple2 = ("apple")     # 쉼표가 없으면 튜플이 아니라 수식이 됨
```

```python
type(single_tuple2)
```
