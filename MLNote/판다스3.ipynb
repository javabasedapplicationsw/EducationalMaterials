### 데이터프레임의 변수별 계산
  - 변수별 합, 평균, 중앙값, 분산, 표준편차 등 통계함수
  - sum(), mean(), median(), var(), std()    

```python
file_url = 'sample_df.csv'
```

```python
sample_df = pd.read_csv(file_url, index_col=0)
```

```python
sample_df
```

```python
sample_df.sum()
```

```python
# 변수별 합과 평균을 출력하시오
# aggregate() 함수의 괄호 안에 sum, mean을 입력하면 됨
# 단, 하나의 값으로 입력되도록 리스트로 묶어주어야 함
sample_df.aggregate(['sum', 'mean'])
```

### iris.csv
  - 붓꽃의 종마다 꽃받침과 꽃잎 크기가 어떻게 다른지를 보여주는 데이터임

```python
file_url = 'iris.csv'
iris = pd.read_csv(file_url)
```

```python
iris
```

```python
'''
class
  - 붓꽃(iris)의 종류
  - versicolor, virginica, setosa
sepal(꽃받침)의 길이와 너비
petal(꽃잎)의 길이와 너비
'''
iris.head()

```

### 가장 우측의 변수 class : 붓꽃(iris)의 종류
### 나머지 4개 변수 :  꽃받침(sepal)과 꽃잎(petal)의 길이와 너비


```python
# 붓꽃 종류별 꽃받침 길이와 너비, 꽃잎 길이와 너비의 평균값을 구하시오.
iris.groupby('class').mean()
```

```python
# 붗꽃의 종류별로 전체 데이터 개수(count)와 평균(mean)을 구하시오.
iris.groupby('class').agg(['count','mean'])
```

```python
# 변수 내 고윳값 확인
# dtype=object   : 텍스트 형태로 된 변수
iris['class'].unique()
```

```python
# nunique() : 몇가지 종류인지 숫자로 확인하기
iris['class'].nunique()
```

```python
# value_counts() : 각 종류별 몇 건의 데이터가 있는지 확인하기
iris['class'].value_counts()
```
