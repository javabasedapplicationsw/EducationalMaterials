# Train / Test 나누기

```python
from sklearn.model_selection import train_test_split
```

```python
X = data.drop('Yearly Amount Spent', axis=1)
```

```python
X
```

```python
y = data['Yearly Amount Spent']
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=100)
```

# 리니어 리그레션 모델 만들기

```python
import statsmodels.api as sm
```

```python
lrm = sm.OLS(y_train, X_train).fit()
```

```python
lrm.summary()
```

# 예측 (평가)

```python
predictions = lrm.predict(X_test)
```

```python
predictions
```

```python
y_test
```

```python
comparison = pd.DataFrame({'actual' : y_test, 'pred' : predictions})
plt.figure(figsize=(10,10))   # 그래프 크기 정의
sns.scatterplot(x='actual', y= 'pred', data=comparison)   # 산점도 그래프 , x축에 실제값 actural, y축에 예측값 pred를 지정함
```
